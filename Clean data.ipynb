# Bộ data bị null ở date

```python
import pandas as pd
data1 = pd.read_csv('/content/drive/MyDrive/Dataset/test_dataset/netflix_titles.csv')
data1
```

```python
data1.info()
```

**Fill date bằng cách lấy giá trị đầu của năm đó rồi fill vào cột date_added**

```python
firt_date = data1['release_year'].astype(str) + '-01-01'
firt_date
```

```python
data1['date_added'].fillna(firt_date, inplace=True)
data1
```

```python
data1['date_added']
```

**Fill cột rating bằng giá trị xuất hiện nhiều nhất**

```python
data1.loc[data1['rating'].isnull()]
```

lấy thể loại xuất hiện nhiều nhất trong rating để fill

```python
fillna_rating = data1['rating'][0]
fillna_rating
```

```python
data1['rating'].fillna(fillna_rating, inplace=True)
```

```python
data1
```

```python
data1.loc[data1['duration'].isnull()]
```

```python
data1['duration'].fillna(data1['rating'], inplace=True)
```

```python
data1.loc[data1['duration'].isnull()]
```

**Tách cột duration thành 2 cột là giá trị và đơn vị**

```python
data1['duration'].str.split(' ', expand=True)
```

```python
data1[['duration_value', 'duration_unit']] = data1['duration'].str.split(' ', expand=True)
data1['duration_unit'] = data1['duration_unit'].str.replace('Seasons', 'Season')
data1
```

**Fill các giá trị bên dưới bằng Unknow**

```python
data1['country'].fillna('Unknow', inplace=True)
data1['cast'].fillna('Unknow', inplace=True)
data1['director'].fillna('Unknow', inplace=True)
data1
```
